```python
def divisors(num):
    y = []
    for number in range(num+1):
        if number != 0:
            if num % number == 0:
                y.append(number)
    return y
```

# CognitiveClass (Data Analysis w/ Python)

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
dc = pd.read_csv(url, header=None)
headers = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors',  'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
dc.columns = headers
dc.head(10)
```

```python
dc.dtypes
```

```python
dc.describe(include = 'all')
```

```python
dc = dc.replace('[?]', np.nan, regex=True)
dc.head(15)
```

```python
#Since we are predicting price, rows that don't have price are useless
dc.dropna(subset = ['price'], axis = 0, inplace = True)
dc.reset_index(drop = True, inplace = True) #Reset index i.e. row number
dc.head(15)
```

```python
dc['normalized-losses'] = dc['normalized-losses'].replace(np.nan, -99)
dc['normalized-losses'] = dc['normalized-losses'].astype('int')
dc.head(15)
```

```python
dc['normalized-losses'] = dc['normalized-losses'].replace(-99, np.nan)
dc.head(15)
```

```python
norm_losses_mean = dc['normalized-losses'].mean()
print (norm_losses_mean)
```

```python
dc.price = dc.price.astype('int')
```

```python
dummy_var = pd.get_dummies(dc['fuel-type'])
dummy_var.tail(15)
```

```python
missing_data = dc.isnull()
missing_data['normalized-losses'].head(15)
```

```python
dc.price = dc.price.astype(float, copy=True)
binwidth = (max(dc.price)-min(dc.price))/4
binwidth
```

```python
bins = np.arange(min(dc.price), max(dc.price)+10070.5, binwidth)
bins
```

```python
group_names = ['Low', 'Medium', 'High', 'Very High']
dc['price-binned'] = pd.cut(dc.price, bins, labels=group_names, include_lowest=True )
dc[['price','price-binned']].head(20)
```

```python
plt.hist(dc.price, bins = 4)

# set x/y labels and plot title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Price bins")
plt.minorticks_on()
```

```python
sns.boxplot(x = 'drive-wheels', y = 'price', data = dc)
```

```python
#Predictor variable on x-axis (Engine Size)
#Target variable on y-axis (Price)
plt.scatter('engine-size', 'price', data = dc)
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.title("Engine Size effect on Price")
plt.minorticks_on()
```

```python
plt.scatter('highway-mpg', 'price', data = dc)
plt.xlabel("Highway mpg")
plt.ylabel("Price")
plt.title("Highway mpg effect on Price")
plt.minorticks_on()
```
